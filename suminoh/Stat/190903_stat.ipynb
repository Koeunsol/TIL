```python
from sklearn.linear_model import LinearRegression # 선형모형
```

```python
model =LinearRegression() # 선형모형 하나를 셋팅 (초깃값을 따로 주지 않아도 됨)
```

```python
x = cars[['speed']]      # 독립변수 
y = cars['dist']            # 종속변수
model.fit(x,y)             # 파라미터 추정   # model에 parameter를 넣어주는 것
```

```python
model.coef_         
# 독립변수의 가중치(ax+b에서 a에 해당)  # codefficient # 계수 #parameter를 표시할 때는 뒤에 _가 붙는 관습이 있음. 
# 독립변수가 여러 개일때도 coef_를 사용하면 여러 개의 계수가 나옴
```

```python
model.intercept_         # 절편(ax+b에서 b에 해당) # 
```

# 실습 : 부트스트래핑으로 선형 모형의 파라미터의 95% 신뢰구간 구하기

```python
coefs = []
intercepts = []

for _ in range(10000):                
    df = resample(cars) # 샘플링
    x = df[['speed']] # 독립변수
    y = df['dist'] # 종속변수
    
    model = LinearRegression() # 모형 만들기
    model.fit(x, y) # 파라미터 추정
    
    coefs.append(model.coef_) # 계수를 리스트에 추가
    intercepts.append(model.intercept_) # 절편을 리스트에 추가
```

```python
coefs
```

```python
intercepts
```

```python
seaborn.distplot(coefs)
```

```python
seaborn.distplot(intercepts)
```

```python
numpy.quantile(coefs, .025)
```

```python
numpy.quantile(coefs, .975)
```

## 데이터 수를 늘리면 신뢰구간이 줄어든다

```python
sample_size = 100
```

```python
ms = []
for _ in range(10000):
    x = uniform(0,100,sample_size)
    m = numpy.mean(x)
    ms.append(m)
numpy.quantile(ms, [.025,.975])
```

## 실습 : sleep 데이터
sleep 데이터를 열어서 group으로 extra를 예측하는 선형 모형을 만듥 각 파라미터의 신뢰구간을 구해보세요.

```python
sleep = pandas.read_csv('sleep.csv')
```

```python
sleep.head()
```

```python
model = LinearRegression()
```

```python
x = sleep[['group']]      # 독립변수 
y = sleep['extra']            # 종속변수
model.fit(x,y)             # 파라미터 추정   # model에 parameter를 넣어주는 것
```

```python
model.coef_
```

```python
model.intercept_
```

```python
seaborn.regplot('group','extra',sleep)
```

```python
coefs = []
intercepts = []

for _ in range(10000):
    df = resample(sleep)
    x = df[['group']]
    y = df['extra']
    
    model = LinearRegression()
    model.fit(x, y)
    
    coefs.append(model.coef_)
    intercepts.append(model.intercept_)
```

```python
seaborn.distplot(coefs)
```

```python
seaborn.distplot(intercepts)
```
