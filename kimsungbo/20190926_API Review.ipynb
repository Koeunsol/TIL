```python
# XML 데이터 파싱
import xml.etree.ElementTree as ET

root = ET.fromstring(response.text)

root
```

```python
# tree에 item 항목만 얻기
item = next(root.iter('item'))

list(item)
```

```python
child = item.find('kindCd')

print(child.text)
```

```python
print(child.tag)
```

```python
# item의 모든자녀
for child in item:
    print(child.tag, child.text)
    
#comprehension
[(child.tag, child.text) for child in item]
```

```python
# 컬럼 목록 얻기
item = next(root.iter('item'))

columns = []

for child in item:
    columns.append(child.tag)
    
columns

#comprehension
columns = [child.tag for child in item]
columns
```

```python
# XML -> DataFrame
rows = []
for item in root.iter('item'):
    row = {}
    for child in list(item):
        row[child.tag] = child.text
    rows.append(row)
    
# comprehension
rows = [{child.tag:child.text for child in list(item)} for item in root.iter('item')]
    
df = pd.DataFrame(rows, columns=columns)
df.sample()
```

## Longitude & Latitude using LocationIQ

```python
def geocode(address):
    url = 'https://us1.locationiq.com/v1/search.php'
    payload = {
        'key': LOCATIONIQ_TOKEN,
        'q': address,
        'format': 'json'
    }
    response = requests.get(url, params=payload)
    results = response.json()
    if 'error' in results:
        return None, None
    coordinates = results[0]
    return coordinates['lat'], coordinates['lon']
```

```python
geocode('서울특별시 구천면로 192')
```

```python
geocode('이 세계 어디에도 없는 주소')
```

```python
sample_df = simple_df.sample(n=500, random_state=1)

sample_df.shape
```

```python
lats = []
lngs = []

for i, index in enumerate(sample_df.index):
    print(i + 1, '-', index)
    lat, lng = geocode(sample_df['orgNm'][index])
    lats.append(lat or 0)
    lngs.append(lng or 0)

sample_df = sample_df.assign(lat=lats, lng=lngs)

sample_df.head()
```

## 지도 그리기    

```python
coordinates = sample_df[['lat', 'lng']]

coordinates
```

```python
map = Map(location=[36.5, 127.5], zoom_start = 7)

for lat, lng in coordinates.values:
    marker = folium.Circle(
        location = [lat, lng],
        radius = 2_000,
        weight = 1,
        fill_color = '#F00',
        fill_opacity = 0.2,
    )
    marker.add_to(map)
    
map
```

```python
# heatmap

map = Map(location=[36.5, 127.5], zoom_start=7)

HeatMap(data=coordinates, radius=20).add_to(map)

map
```
