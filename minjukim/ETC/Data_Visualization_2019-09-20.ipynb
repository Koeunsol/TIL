# 1. Matplotlib
Customizing 하기 좋다

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline # matplotlib 결과물을 바로 주피터노트북에서 볼 수 있게 해준다
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel('ind var')
plt.ylabel('dep var')
plt.show()
```

```python
plt.subplot(1, 2, 1) # 1 x 2 Matrix
plt.plot(x, y)
plt.title('y = x^2')

plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.title('y = x^0.5')
```

```python
plt.subplot(2, 1, 1) # 2 x 1 Matrix
plt.plot(x, y)
plt.title('y = x^2')

plt.subplot(2, 1, 2)
plt.plot(y, x)
plt.title('y = x^0.5')
```

# 객체지향 프로그래밍 Object-Oriented Programming

Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 2, 1]) 
# [전체 캔버스 내에서 그래프가 시작되는 위치 좌표, 그림의 크기 좌표]
axes.plot(x, y)
axes.set_title('y = x^2')
axes.set_xlabel('ind var')
axes.set_ylabel('dep var')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_title('y = x^2')

axes2 = fig.add_axes([0.1, 0.7, 0.2, 0.2])
axes2.plot(y, x)
axes2.set_title('y = x^0.5')
```

### Object-Oriented : Canvas, Axes 를 미리 그려놓고 위치정보를 꺼내 쓴다

```python
fig, axes = plt.subplots(1, 2) 
# fig 에는 전체 캔버스가 저장된다
# axes 에는 생성된 그래프의 array 가 Tuple 형태로 저장된다
```

```python
axes[0].plot(x, y)
axes[1].plot(y, x)
fig
```

# 범례Legend 지정

```python
# 그래프 2 개 그리기
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='y = x^2')
ax.plot(x, x**3, label='y = x^3')
ax.legend(loc=0) # Legend 함수를 써서 불러내야 그래프 상에 표시된다
                 # loc=0 : default (1, 2, 3, 4 ... 10(센터) 바꿀 수 있다)
                 # loc=[0, 0] : 리스트 형태로 좌표 할당 가능, 그래프 밖으로 내보낼 수 있다
```

```python
plt.figure()
plt.plot(x, y, 'b.--') # r: red, y: yellow, b: blue, g: green 
                       # o, x, p, s, *, .: marker shape
                       # --: 추세선

plt.plot(x, y, color = 'black', marker = 'p', ms = 20, ls = '-.', lw = 3)        
plt.plot(x, y, color = '#088888', marker = 'p', ms = 20, ls = '-.', lw = 3)
```

# Relation Plot

```python
sns.relplot(x='total_bill', y='tip', data=tips, col='sex', hue='smoker', style='smoker', size='size')
```

```python
time = np.arange(500)
value = np.random.randn(500).cumsum() # .cumsum() 누적 합
```

```python
df = pd.DataFrame(dict(time = time, value = value))
df.sample()
```

```python
# relplot: relativity plot
sns.relplot(x = 'time', y = 'value', kind='line', data=df)

# sns.relplot(x = 'time', y = 'value', data=df)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri)
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', hue='event', style='event', data=fmri) # 흐름의 트렌드를 가장 잘 보여준다
sns.relplot(x='timepoint', y='signal', kind='line', hue='event', style='event', ci=None, data=fmri) # ci: confidence interval신뢰구간
sns.relplot(x='timepoint', y='signal', kind='line', hue='event', style='event', estimator=None, data=fmri) # estimator: 추정값 없이 보여줘라
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# px 안에 있는 데이터 불러오기
iris = px.data.iris()
iris.sample()
```

```python
px.scatter(iris, x='sepal_width', y='sepal_length', color='species', size='petal_length')
```

```python
data = px.data.gapminder()
data.sample()
```

```python
data['country'].unique()
```

```python
canada = data[data['country'] == 'Canada']
canada.sample()
```

```python
brazil = data[data['country'] == 'Brazil']
brazil.sample()
```

```python
px.bar(brazil, x='year', y='pop', color='gdpPercap')
```

```python
px.line(data, x='year', y='lifeExp')
```

```python
oceania = data.query("continent =='Asia'")
# oceania
px.line(oceania, x='year', y='lifeExp', color='country')
```

```python
# 주가 데이터
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/stockdata2.csv')
df.sample()
```

```python
msft = df[df['stock'] == 'MSFT']
msft.sample()
```

```python
px.line(msft, x='Date', y='value')
```

# MicroSoft 와 Apple 의 주가를 비교해서 한 눈에 보고싶다

```python
aapl = df[df['stock'] == 'AAPL']
aapl.sample()
```

```python
ibm = df[df['stock'] == 'IBM']
ibm.sample()
```

```python
sbux = df[df['stock'] == 'SBUX']
sbux.sample()
```

```python
gspc = df[df['stock'] == 'GSPC']
gspc.sample()
```

```python
fig = go.Figure() # Canvas 를 먼저 그린다
fig.add_trace(go.Scatter(x=msft['Date'], y=msft['value'], mode='lines', name='MSFT')) # 위에서 그린 px 와 똑같은 그래프를 그린 것
fig.add_trace(go.Scatter(x=aapl['Date'], y=aapl['value'], mode='lines', name='AAPL'))
fig.add_trace(go.Scatter(x=ibm['Date'], y=ibm['value'], mode='lines', name='IBM'))
fig.add_trace(go.Scatter(x=gspc['Date'], y=gspc['value'], mode='lines', name='GSPC'))
fig.add_trace(go.Scatter(x=sbux['Date'], y=sbux['value'], mode='lines', name='SBUX'))
```
