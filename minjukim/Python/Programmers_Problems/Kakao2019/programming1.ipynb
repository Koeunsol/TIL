```python
# 문제는 3개단위 끊고 남은 2개 등등 처리
# 그다음에 각각 끊어서 한 것 중에 제일 작은 값 리턴
def solution(s):
    answer = []
    
    # 한개
    for i in range(len(s)//1):
        if i == 0:
            answer.append(s[i])
        elif s[1*(i-1):1*1] == s[1*i:1*(i+1)]:
            if type(answer[-1]) == int:
                answer.append(answer[-1] + 1)
                answer.pop(-2)
            else:
                answer.append(2)

        else:
            answer.append(s[1*i])
#             answer.append(s[1*i+1])
        continue
        answer.append(s[i])

    # 두개
    for i in range(len(s)//2):
        if i == 0:
            answer.append(s[i])
            answer.append(s[i+1])
        elif s[2*(i-1):2*i] == s[2*i:2*(i+1)]:
            if type(answer[-1]) == int:
                answer.append(answer[-1] + 1)
                answer.pop(-2)
            else:
                answer.append(2)
        else:
            answer.append(s[2*i])
            answer.append(s[2*i+1])
#             answer.append(s[2*i+2])
            
    # 세개 
    for i in range(len(s)//3):
        if i == 0:
            answer.append(s[i])
            answer.append(s[i+1])
            answer.append(s[i+2])
        elif s[3*(i-1):3*i] == s[3*i:3*(i+1)]:
            if type(answer[-1]) == int:
                answer.append(answer[-1] + 1)
                answer.pop(-2)
            else:
                answer.append(2)
        else:
            answer.append(s[3*i])
            answer.append(s[3*i+1])
            answer.append(s[3*i+2])
            answer.append(s[3*i+3])            
            
            
    return len(answer)
```
