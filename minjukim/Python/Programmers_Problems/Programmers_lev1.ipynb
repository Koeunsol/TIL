# 행렬의 덧셈

```python
def solution(arr1, arr2):
    num_row = len(arr1)
    num_col = len(arr1[0])
    
    answer = []
    row = []
    
    for r in range(num_row):
        for c in range(num_col):
            element = arr1[r][c] + arr2[r][c]
            row.append(element)

        answer.append(row)
        row = []
        
    return answer
```

```python
solution([[1,2],[2,3]], [[3,4],[5,6]])
```

```python
assert solution([[1,2],[2,3]], [[3,4],[5,6]]) == [[4,6],[7,9]]
assert solution([[1],[2]], [[3],[4]]) == [[4],[6]]
```

```python
# 짧은 코드
def sumMatrix(A,B):
    answer = [[c + d for c, d in zip(a, b)] for a, b in zip(A,B)]
    return answer
```

# 이상한 문자 만들기

```python
def solution(s):
    answer = []
    for word in s.split():
        for i, x in enumerate(word):
            if i % 2 == 0:
                answer.append(x.upper())
            else : answer.append(x.lower())
        answer.append(' ')    # 하나 이상의 공백문자
    return ''.join(answer[:-1])
```

```python
def solution(s):
    answer = ''
    count = 0
    
    for letter in s:
        if letter == ' ':
            count = -1

        else: 
            if count % 2 == 0:
                letter = letter.upper()
            else: letter = letter.lower()
                
        answer += letter
        count += 1

    return answer
```

```python
# solution('try hello world')
solution('try  hello  world')
```

```python
assert solution('try hello world') == 'TrY HeLlO WoRlD'
```

# 자연수 뒤집어 배열로 만들기

```python
def solution(n):
    answer = []
    
    while n >= 10:
        answer.append(n % 10)
        n = n // 10
    answer.append(n)
    
    return answer
```

```python
solution(12345)
```

```python
# 개짧은 코드
def digit_reverse(n):
    return list(map(int, reversed(str(n))))
```

# 제일 작은 수 제거하기

```python
def solution(arr):
    if len(arr) == 1: return [-1]
    arr.remove(min(arr))
    return arr
```

```python
solution([4,3,2,1])
```

# 최대공약수와 최소공배수

```python
def solution(n, m):

    for i in range(1, min(n, m) + 1):
        if (n % i == 0) & (m % i == 0):
            greatest_common = i
    
    i = min(n, m)            
    while i:
        if (i % n == 0) & (i % m == 0):        
            least_common = i
            break
        i += 1
            
    return [greatest_common, least_common]
```

```python
solution(3, 12)
```

# 하샤드 수

```python
def solution(x):
    return x % sum([int(i) for i in str(x)[::-1]]) == 0
```
