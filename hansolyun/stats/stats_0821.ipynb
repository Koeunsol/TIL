## 모수 개념 설명

### 모수(parameter->설정값)
- 분포를 설정해주는 값이 parameter
- 동전을 10개 던질 때 모수는 10 -> 분포의 최댓값은 죽었다 깨어나도 10이기 때문에 10(모수)가 분포를 설정한다
- 설정을 잡아주는 값
- 기하분포에서는 확률이 parameter
- 모수를 레시피라고 생각하면 됨 -> 레시피가 케익맛을 결정하지만 몇개 만드는지는 레시피를 결정하지 않기때문에

### 분포 
- 데이터의 모양

## 실습
- 골든 리트리버 예제

```python
개판 = []
for i in range(100):
    한계 = 10  #parameter 1
    옐로카드 = 0    #parameter 2
    짜증확률 = .4
    기간 = 0
    while 옐로카드 < 한계: 
        짜증 = binomial(1, 짜증확률)
        if 짜증 == 1:
            옐로카드 += 1 #옐로카드 = 옐로카드 +1
        else:
            옐로카드 = 0
        기간 += 1
    개판.append(기간)
sb.distplot(개판)
```

## 확률
- 확률 P(A)
- 결합 확률 P(A,B)
- 조건부 확률 P(B|A) = P(A,B)/P(A)

```python
acting = normal(50, 15, 100) #평균이 50, 표준편차 15인 배우 100명
sb.distplot(acting)
```

```python
average_revenue = 100 * acting + 20000
```

```python
revenue = normal(average_revenue, 800) #각 평균매출에 표준편차는 8000만원으로 동일
```

```python
sb.distplot(revenue)
```

```python
sb.jointplot(acting, revenue)
```

```python
sb.regplot(acting, revenue)
```
